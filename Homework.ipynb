In your Python script, import matplotlib.pyplot, DecisionTreeClassifier, and plot_tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

Create Your Dataset:
Encode your choices for snacks into a simple dataset. Use numerical representations for choices.

```python
X = [[0, 0],  # Sweet -> Apple
     [1, 2]]  # Salty -> Crackers
y = [0, 2]    # Final Choice: Apple=0, Crackers=2
```

Train the Decision Tree:
Create a DecisionTreeClassifier instance and fit it to your dataset.

```python
clf = DecisionTreeClassifier(random_state=123)
clf.fit(X, y)
```

Visualize the Decision Tree:
Use plot_tree from scikit-learn to visualize the trained decision tree.

```python
plt.figure(figsize=(8,6))
plot_tree(clf, filled=True, feature_names=['First Choice', 'Second Choice'], class_names=['Apple', 'Crackers'])
plt.show()
```

Make a Prediction:
Create a new observation and use the decision tree to make a prediction.

```python
new_observation = [[0, 1]]  # Example: Sweet -> ???
prediction = clf.predict(new_observation)
```

Display the Prediction:
Map the prediction back to snack names and print the result.

```python
snacks = {0: 'Apple', 2: 'Crackers'}
predicted_snack = snacks.get(prediction[0], "Unknown Snack")
print(f"The decision tree has predicted that the ideal snack choice would be: {predicted_snack}")
```
